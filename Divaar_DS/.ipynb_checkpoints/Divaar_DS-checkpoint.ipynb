```python
cities = data.city.value_counts()
```

```python
import matplotlib.pyplot as plt
plt.bar(cities.keys(), cities.values)
plt.xlabel('city')
plt.ylabel('Number of posts')
plt.xticks(rotation=50)
plt.show()
```

```python
cat1 = data.cat1.value_counts()
plt.bar(cat1.keys(), cat1.values)
plt.xlabel('cat1')
plt.ylabel('Number of posts')
plt.xticks(rotation=50)
plt.show()
```

```python
subd1 = data[data.cat1 == 'for-the-home']
cat = subd1.cat2.value_counts()
plt.bar(cat.keys(), cat.values)
plt.xlabel('cat2')
plt.ylabel('Number of posts')
plt.xticks(rotation=50)
plt.show()
```

```python
subd2 = data[(data.cat1 == 'for-the-home') & (data.city=='Qom')]
cat = subd2.cat2.value_counts()
plt.bar(cat.keys(), cat.values)
plt.xlabel('cat2')
plt.ylabel('Number of posts')
plt.xticks(rotation=50)
plt.show()
```

```python
subd2 = data[(data.cat1 == 'for-the-home') & (data.city=='Tehran')]
cat = subd2.cat2.value_counts()
plt.bar(cat.keys(), cat.values)
plt.xlabel('cat2')
plt.ylabel('Number of posts')
plt.xticks(rotation=50)
plt.show()
```
