```python
#remove number and punctuations:
sample_data.comment = sample_data.comment.map(lambda x : re.sub('\d+', '', x))
sample_data.comment = sample_data.comment.map(lambda x : x.translate(str.maketrans('', '', string.punctuation)) )
```

```python
sample_data.recommend.iloc[0] =='\\N'
```

```python
sample_data2 = sample_data[sample_data.recommend != '\\N']
```

```python
sample_data2 = sample_data2[['recommend', 'comment']]
```

```python
sample_data2.head()
```

```python
sample_data2.comment = sample_data2.comment.map(lambda x:re.sub('\n', ' ', x))
```

```python
words_list = []
for comment in sample_data2.comment:
    words_list += word_tokenize(comment)
```

```python
len(words_list)
```

```python
counts = Counter(words_list)
```

```python
vocab = sorted(counts, key=counts.get, reverse=True)
vocab_to_int = {word: ii for ii, word in enumerate(vocab, 1)}
```
